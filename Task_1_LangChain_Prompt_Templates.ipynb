# **TASK-5:Input Validation Layer**

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience. Default set to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Default set to 'formal'")
        tone = "formal"

    return audience, tone
```

```python
audience = input("Enter audience: ")
tone = input("Enter tone: ")

# Validate inputs
audience, tone = validate_inputs(audience, tone)

# Print final values
print("\nFinal Values:")
print("Audience:", audience)
print("Tone:", tone)
```

# **TASK-6:Propmt Generator App**

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    templates = {

        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} step by step for {audience} in a {tone} tone"
        ),

        "interview": PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 interview questions about {topic}"
        ),

        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    }

    selected_template = templates.get(style)

    if style == "interview":
        return selected_template.format(topic=topic)

    return selected_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_prompt(
    "Neural Networks",
    "beginner",
    "fun",
    "storytelling"
))
```

# **Task-7: Template Reusability Test**

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "formal"),
    ("Python", "student", "fun"),
    ("Data Science", "expert", "casual"),
    ("Machine Learning", "intermediate", "formal"),
    ("Deep Learning", "researchers", "technical")
]

for topic, audience, tone in inputs:
    print(reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
```
